```python
import pandas as pd 
import numpy as np 
```

```python
transport = pd.read_csv("../sources/transportation/passengers63-68.csv")
```

```python
transport.isna().sum()
```

```python
transport['ยานพาหนะ/ท่า'].unique()
```

```python
transport['ยานพาหนะ/ท่า'] = transport['ยานพาหนะ/ท่า'].replace({
    'รถไฟฟ้าสายสีชมพูู': 'รถไฟฟ้าสายสีชมพู',  # แก้ไขการพิมพ์ผิด
    'รถไฟสายสีแดง': 'รถไฟฟ้าสายสีแดง'  # ปรับให้เป็นรูปแบบเดียวกัน
})
```

```python
# ตรวจสอบค่าที่มีในคอลัมน์
print(transport['ยานพาหนะ/ท่า'].unique())

# กรองเฉพาะข้อมูลรถไฟฟ้า
allowed_rail = [
    'รถไฟฟ้าสายสีน้ำเงิน',
    'รถไฟฟ้าสายสีม่วง',
    'รถไฟฟ้า ARL',
    'รถไฟฟ้า BTS',
    'รถไฟฟ้าสายสีแดง',
    'รถไฟฟ้าสายสีเหลือง',
    'รถไฟฟ้าสายสีชมพู'
]
transport = transport[transport['ยานพาหนะ/ท่า'].isin(allowed_rail)].reset_index(drop=True)

# ตรวจสอบค่าหลังการกรอง
transport['ยานพาหนะ/ท่า'].unique()
```

```python
transport.drop(columns=['รูปแบบการเดินทาง', 'วัตถุประสงค์', 'สาธารณะ/ส่วนบุคคล', 'หน่วยงาน', 'หน่วย'], inplace=True)
```

```python
transport[transport['ยานพาหนะ/ท่า'] == 'รถไฟฟ้าสายสีชมพู'] 
```

```python
transport.to_csv('../sources/transportation/daily_transport_rail.csv', index = False)
```
